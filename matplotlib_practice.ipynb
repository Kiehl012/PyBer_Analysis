```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2)
ax.legend('Boston')
ax.set_title('Pyber Fares By Month')
ax.set_xlabel('Month')
ax.set_ylabel('Fare ($)')
ax.set_ylim(5,45)
ax.grid()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.figure(figsize=(10,10))
plt.legend('Boston')
plt.xlabel('month')
plt.ylabel('fare($)')
plt.title('Pyber Fares')
plt.gca().invert_yaxis()
plt.xlim(0,35)
plt.scatter(y_axis, x_axis, c='red')
```

```python
y_axis_larger = [item*3 for item in y_axis]
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
y_axis_bigger = [item*5 for item in y_axis]
ax.scatter(y_axis, x_axis, s=y_axis_bigger, c='skyblue', alpha=.8, edgecolor='black',linewidths=2)
ax.set_title('Pyber Fares by Month')
ax.set_xlim(0,60)
ax.invert_yaxis()
ax.legend(['Boston'])
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%', textprops={'color':"w"})
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%',
    textprops={'color':'w'})

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,10))
explode_values = [0,0,0.2,0,0,0,0.2,0,0,0,0,0]
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
ax.pie(y_axis,labels=x_axis, textprops={'color':'w'}, explode=explode_values, autopct="%.1f%%", startangle =-100, colors=colors, shadow=True)
ax.invert_yaxis()
plt.show()
```
